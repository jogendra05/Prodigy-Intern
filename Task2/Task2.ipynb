```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'Movies1.csv')
df.dropna(inplace = True)
df.replace('₹', '', regex= True, inplace=True)
df.replace({
    'Budget (est.)': '[A-Za-z]',
    'World wide gross (INR)': '[A-Za-z]',
    'World wide gross (USD)': '[A-Za-z]',
    'Gross in India (INR crore)': '[A-Za-z]'
}, '', regex= True, inplace=True)
df.replace({
    'Budget (est.)': '\['
}, '', regex = True, inplace=True)
df
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df['Budget (est.)'] = pd.to_numeric(df['Budget (est.)'], errors = 'coerce').fillna(0).astype(int)
df['World wide gross (INR)'] = pd.to_numeric(df['World wide gross (INR)'], errors ='coerce').fillna(0).astype(int)
df['World wide gross (USD)'] = pd.to_numeric(df['World wide gross (USD)'], errors = 'coerce').fillna(0).astype(int)
df['Gross in India (INR crore)'] = pd.to_numeric(df['Gross in India (INR crore)'], errors = 'coerce').fillna(0).astype(int)
df
```

```python
df.describe()
```

```python
new_df = df.set_index('Film')
new_df
```

```python
new_df = df.drop(columns = ['Year']) 
new_df.sort_values('World wide gross (INR)',ascending = False, inplace=True)
new_df = new_df.head()
new_df.plot(kind= 'bar',y = 'World wide gross (INR)', x = 'Film', figsize=(15,7))
plt.title('Top 5 Movies', fontsize=16)
plt.xlabel('Film', fontsize=14)
plt.ylabel('INR', fontsize=14)
plt.legend(fontsize = 12)
plt.xticks(rotation = 0)
for i, v in enumerate(new_df['World wide gross (INR)']):
    plt.text(i, v+1, str(v), ha= 'center', va= 'bottom')
```

```python
Profit_movies = df[df['Budget (est.)'] <= df['World wide gross (INR)']]
Profit_movies
```

```python
profit = 0
for index, row in df.iterrows():
    if row['Budget (est.)'] <= row['World wide gross (INR)']:
        profit += 1
print(f"From {len(df)} Movies, {profit} are profit making and {len(df) - profit} are loss making")
```

```python
df['Profit_%'] = ((df['World wide gross (INR)'] - df['Budget (est.)'])/df['Budget (est.)'])*100
max_profit = df[df['Profit_%'] == df['Profit_%'].max()]
max_profit[['Film', 'Year', 'Profit_%']]
```
