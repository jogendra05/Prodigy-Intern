```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'Accident_Information.csv')
df
```

```python
df.isnull().sum()
```

```python
df.drop(['Accident_Index', '1st_Road_Class', '2nd_Road_Class'], axis = 'columns', inplace = True)
```

```python
df.dropna(inplace = True)
```

```python
df.isnull().sum()
```

```python
df = df.set_index('Date')
df = df.head(1000)
df
```

```python
df.describe()
```

```python
df.describe(include = 'object')
```

```python
severity = df['Accident_Severity'].value_counts()
severity
```

```python
plt.pie(severity, labels = severity.index, autopct = '%1.1f%%') 
plt.title('Distribution of Accident Severity')
```

```python
new_df = df.groupby('Day_of_Week')['Number_of_Vehicles'].sum()
new_df
```

```python
new_df.plot(kind='bar', x = 'Day_of_Week', y = 'Number_of_Vehicles', figsize= (10,7)) 
plt.title('No. of Accident on each Day')
plt.xlabel('Days')
plt.ylabel('No. of accidents')
plt.xticks(rotation = 0)
```

```python
weather = df['Road_Surface_Conditions'].value_counts()
weather
```

```python
plt.pie(weather, labels = weather.index, autopct= '%1.1f%%')
plt.title('Distribution of Accident in terms of Weather')
```

```python
road = df['Road_Surface_Conditions'].value_counts()
plt.pie(road, labels = road.index, autopct = '%1.1f%%')
plt.title('Condition of Roads')
```

```python
import geopandas as gpd
```

```python
india = gpd.read_file(r'C:\Users\Jogendra\Downloads\data\polbnda_ind.shx')
gdf = gpd.GeoDataFrame(df, geometry=gpd.points_from_xy(df.Longitude, df.Latitude))

fig, ax = plt.subplots(figsize=(10, 10))
india.plot(ax=ax, color='lightgrey')
gdf.plot(ax=ax, color='red', markersize=50, label='Accidents', marker='o')
plt.title('Accidents in India')
plt.legend()
```
