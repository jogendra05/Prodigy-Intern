```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'population_by_age_group.csv')
df
```

```python
df1 = df.sort_values('Total', ascending = False)
df1
```

```python
df = df1.set_index('Country').head()
df
```

```python
df.describe()
```

```python
new_df = df.drop(columns=['Total'])
new_df.plot(kind='bar', figsize=(15, 5), width = 0.6)
plt.xlabel('Country', fontsize =14)
plt.ylabel('Age', fontsize =14)
plt.title('Top 5 Countries Age Distribution', fontsize =16)
plt.legend(fontsize=12)
plt.xticks(rotation = 0)
```

```python
df.plot(kind='bar', y = 'Total', figsize=(10, 6))
plt.xlabel('Country', fontsize =14)
plt.ylabel('Age', fontsize =14)
plt.title('Top 5 Countries Population', fontsize =16)
plt.legend(fontsize=12)
plt.xticks(rotation = 0)
for i, v in enumerate(df.Total):
    plt.text(i, v+1, str(v), ha='center', va='bottom')
```

```python
df2 = df1.tail()
df2.set_index('Country', inplace = True)
low = df2.drop(columns='Total')
low.plot(kind='bar', figsize = (15, 5))
plt.xlabel('Country', fontsize=14)
plt.ylabel('Age', fontsize=14)
plt.title('Least 5 Countries Age Distribution', fontsize=16)
plt.legend(fontsize=12)
plt.xticks(rotation = 0)
```

```python
df2.plot(kind = 'bar', y = 'Total', figsize=(10, 6))
plt.xlabel('Country', fontsize = 14)
plt.ylabel('Age', fontsize = 14)
plt.title('Least 5 Populated Countries', fontsize = 16)
plt.legend(fontsize = 12)
plt.xticks(rotation = 0)
for i, v in enumerate(df2.Total):
    plt.text(i, v+1, str(v), ha ='center', va='bottom')
```
