```python
from sklearn.preprocessing import LabelEncoder
# from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
```

```python
object.columns.tolist()
la_encoder = LabelEncoder()
for col in object:
    df[col] = la_encoder.fit_transform(df[col])
df
```

```python
scaler = MinMaxScaler()
scaler.fit(df[['age', 'duration', 'campaign', 'previous', 'emp.var.rate', 'cons.price.idx', 'cons.conf.idx']])
df[['age', 'duration', 'campaign', 'previous', 'emp.var.rate', 'cons.price.idx', 'cons.conf.idx']] = scaler.transform(df[['age', 'duration', 'campaign', 'previous', 'emp.var.rate', 'cons.price.idx', 'cons.conf.idx']])
df
```

```python
input = df.drop('y', axis='columns')
input
```

```python
target = df['y']
target
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
model.fit(input, target)
```

```python
model.score(input, target)
```

```python
# Predicting if it gives 0 then customer will not purchase the product 
model.predict([[0.41, 4,1,4,0,0,0.22,0.03,0,1,0.68,0.36,0.34]])
```

Here there are less chances for buying a service or product

```python
# Predicting if it gives 1 then customer will or can purchase the product 
model.predict([[0.5,1,1,0,0,2,0.77,0,0,1,1,0.69,0.31]])
```

Here there is a high possibilites for buying a product or service
